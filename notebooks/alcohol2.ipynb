```python
import pandas as pd
pd.set_option('display.max_columns', 50)
```

```python
state_pop = pd.read_csv("../data/state-pops.csv")
```

```python
state_pop = state_pop[:52]
state_pop
```

```python
state_pop['Census population, April 1, 2010\n[3]'] = state_pop['Census population, April 1, 2010\n[3]'].str.replace(r',', '').astype(str)
```

```python
state_pop['Census population, April 1, 2010\n[3]']=state_pop['Census population, April 1, 2010\n[3]'].astype(float)
```

```python
type(state_pop['Census population, April 1, 2010\n[3]'][0])
```

```python
pd.read_csv("../output/alc_suicide_states.csv")
```

```python
state_suicide = pd.read_csv("../output/alc_suicide_states.csv")
```

```python
merged = state_suicide.merge(state_pop, left_on = "LocationDesc", right_on = "State")
```

```python
merged.head()
```

```python
merged["Rate"]  = merged["Data_Value_Alt"]*100000/merged["Census population, April 1, 2010\n[3]"]
```

```python
merged.head()
```

```python
suicide_rates1 = merged[['LocationDesc', 'Rate']]
```

```python
suicide_rates1
```

```python
suicide_rates1.sort_values("Rate", ascending = False)
```

```python
suicide_rates2 = suicide_rates1.sort_values("Rate", ascending = False)
```

```python
suicide_rates2.to_csv("../output/final_suicide_bystate.csv")
```
